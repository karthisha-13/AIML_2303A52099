```python
people = ['karthisha','sowmya','sathwika','sushanthi','navya','vyshu']
age = [20,21,22,40,18,22,]
weight = [55,35,77,68,70,60]
height = [160,135,170,165,173,168]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize=16,
          fontstyle='italic')

# Main plot function 'bar'
plt.bar(x=people,height=weight, width=0.6,
        color='pink',edgecolor='k',alpha=0.8)

plt.xlabel("People",fontsize=10)
plt.xticks(fontsize=12,rotation=25)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))

# Main plot function 'hist'
plt.hist(weight,color='blue',edgecolor='k', alpha=0.75,bins=5)

plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
